```python
plt.figure(figsize=(30,5))
sns.barplot(data = socialMediaData, x = 'Primary Platform', y = 'Gender')
plt.xlabel('Primary Platform')
plt.ylabel('Gender')
plt.title('Platform per Gender')
plt.show()
```

```python
sns.pairplot(socialMediaData, x_vars=['Country','Occupation'], 
             y_vars=['Primary Platform'], kind='scatter')
```

```python
gender_social_time = socialMediaData.groupby("Gender")["Daily Social Media Time (hrs)"].mean()
print(gender_social_time)
```

```python
occupation_screen_time = socialMediaData.groupby("Occupation")["Screen Time (hrs)"].mean()
occupation_screen_time
```

```python
platform_usage = socialMediaData.groupby("Primary Platform")["User ID"].count()
print(platform_usage)
```

```python
country_summary = socialMediaData.groupby("Country").agg(
    avg_screen_time=("Screen Time (hrs)", "mean"),
    avg_income=("Monthly Income (USD)", "count"),  # you may need to convert income to numeric first
    avg_sleep=("Average Sleep Time (hrs)", "mean")
)
country_summary
```

```python
tech_vs_gaming = socialMediaData.groupby("Tech Savviness Level (scale 1-10)")["Daily Gaming Time (hrs)"].mean()
tech_vs_gaming
```
